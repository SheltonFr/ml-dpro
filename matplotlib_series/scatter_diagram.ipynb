# Scatter Diagram

```python
import matplotlib.pyplot as plt

science_score = [43, 78, 73, 86, 89, 61, 84, 66, 87, 89]
math_score = [51, 71, 69, 82, 80, 60, 88, 56, 85, 99]

fig, ax = plt.subplots()
ax.scatter(science_score, math_score)
plt.show()
```

## Chenge marker style

```python
fig, ax = plt.subplots()
ax.scatter(science_score, math_score, marker='x')
plt.show()
```
