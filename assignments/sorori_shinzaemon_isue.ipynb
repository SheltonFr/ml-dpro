## [Probme 1] Number of grains of rice on the 100th day

```python
def calculate_total_number_of_grains(days: int = 100) -> float:
    current_day = 0
    total = 0
    for num in range(days):
        if num == 0:
            current_day = 1
        else:
            current_day = current_day * 2
        total = total + current_day
        
    return total
```

### Line Graph

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 101)
y = [calculate_total_number_of_grains(num) for num in x]

plt.plot(x,y)
plt.title("B")
plt.xlabel("Number of Days")
plt.ylabel("Grains of Rice")
```

## [Problem 2] Number of rice grains outside of the 100th day

Once the previous function cover all cases for this specific problem, we can pass as parameter any value, so we'll know the answer for any possible day.  

Lets use 200 days as example

```python
number_of_grains_in_200days = calculate_total_number_of_grains(200)
```

```python
print(f'{number_of_grains_in_200days} grains of rice')
```
