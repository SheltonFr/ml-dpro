## Line Graph

### Draw multiple polygonal lines

```python
import matplotlib.pyplot as plt
```

```python
x1 = [1, 2, 3]
y1 = [2, 4, 6]
x2 = [1, 2, 3]
y2 = [6, 2, 4]
```

__fig__: fornece metodos para manipular o estilo  
__ax__: fornece metodos para o esboco

```python
fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.plot(x2, y2)
```

## Curve Graph

```python
import numpy as np

# an array from 0 - 99 with step 1
x = np.arange(0, 100, 1)
y = np.arange(0, 20, 0.2)
yy = y*y

fig, ax = plt.subplots()
ax.plot(x, yy)
plt.show()
```
